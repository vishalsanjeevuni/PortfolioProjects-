```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#load in the data
df = pd.read_csv("C:/Users/vishal/Desktop/Professional Stuff/Projects/Portfolio projects data/Visualizing Categorical Data/exploring_mushrooms.csv")
print(df.head(5))
```

```python
#list of all column headers
columns = df.columns.tolist()
print(columns)
```

```python
for column in columns:  
 sns.countplot(x=df[column], order=df[column].value_counts().index)
 plt.title(column + " Value Counts")
 plt.show()
 plt.clf()
 # rotates the value labels slightly so they dont overlap, also slightly increases font size
 plt.xticks(rotation=30, fontsize=10)
 # increases the variable label font size slightly to increase readability
 plt.xlabel(column, fontsize=12)
```
