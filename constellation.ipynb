Constellation Orion, a collection of stars that appear in our night sky and form the shape of [Orion](https://en.wikipedia.org/wiki/Orion_(constellation)), a warrior God from ancient Greek mythology. 

visualizing the Orion constellation in 3D using the Matplotlib.
Created a rotate-able visualization of the position of the Orion's stars and get a better sense of their actual positions. To achieve this, I have mapped a real data from outer space that maps the position of the stars in the sky

The goal of the project is to understand spatial perspective. To visualize Orion in both 2D and 3D, to see the difference in the constellation shape humans see from earth versus the actual position of the stars that make up this constellation.

## 1. Set-Up

+ Adding `%matplotlib notebook` to rotate visualization.

+ In order to see 3D visualization: `from mpl_toolkits.mplot3d import Axes3D`


```python
%matplotlib notebook
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

## 2. About data

Astronomers describe a star's position in the sky by using a pair of angles: declination and right ascension. Declination is similar to longitude, but it is projected on the celestian fear. Right ascension is known as the "hour angle" because it accounts for time of day and earth's rotaiton. Both angles are relative to the celestial equator. To learn more about star position [here](https://en.wikipedia.org/wiki/Star_position).

The `x`, `y`, and `z` lists below are composed of the x, y, z coordinates for each star in the collection of stars that make up the Orion constellation as documented in a paper by Nottingham Trent Univesity on "The Orion constellation as an installation" found [here](https://arxiv.org/ftp/arxiv/papers/1110/1110.3469.pdf).

```python
# Orion
x = [-0.41, 0.57, 0.07, 0.00, -0.29, -0.32,-0.50,-0.23, -0.23]
y = [4.12, 7.71, 2.36, 9.10, 13.35, 8.13, 7.19, 13.25,13.43]
z = [2.06, 0.84, 1.56, 2.07, 2.36, 1.72, 0.66, 1.25,1.38]
```

## 3. Creating a 2D Visualization

Used the scatter function to visualize `x` and `y` coordinates.

There is a curve to the sky, and this is a flat visualization.

```python
plt.figure()
ax= plt.subplot(1,1,1)
plt.scatter(x,y)
plt.title("2D visualization of the Orion constellation")
plt.show()
```

## 4. Creating a 3D Visualization

visualizing in 3D in the to get a better sense of the actual star positions. 


```python
fig_3d= plt.figure()
ax= plt.subplot(1,1,1,projection="3d")
constellation3d= plt.scatter(x,y,z)
plt.title("3D visualization of the Orion constellation")
plt.show()
```

## 5. Summary

Use your mouse to click and drag the 3D visualization. This will rotate the scatter plot. As you rotate, You can see Orion from different angles.

The visualization does not curve as the night sky does.
This 3D visualization helps understanding of Earthly perspective! We see the shape of the warrior Orion because of Earth's location in the universe and the location of the stars in that constellation.
